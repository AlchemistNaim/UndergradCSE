### **Semester 1: Foundation in AI/ML**
- **Coursework:**
  - Focus on core computer science subjects while selecting electives in AI/ML (e.g., Introduction to Machine Learning, Data Science).
  - Take a course in **statistics** and **probability**, as they are foundational for AI/ML.

- **Skills Development:**
  - Learn Python and essential libraries like NumPy, pandas, and Scikit-learn.
  - Start practicing on platforms like **Kaggle** or **LeetCode** to build problem-solving skills.

- **Project:**
  - Begin a small AI/ML project, such as a simple classifier or regression model. Document this project well, as it will be useful for your portfolio.

### **Semester 2: Advanced AI/ML Concepts**
- **Coursework:**
  - Enroll in advanced AI/ML courses (e.g., Deep Learning, Natural Language Processing).
  - Take a course related to **databases** or **data engineering** to understand data handling.

- **Skills Development:**
  - Dive into deep learning frameworks like **TensorFlow** or **PyTorch**.
  - Learn about **version control** (Git) and basic software development practices.

- **Project:**
  - Work on a more complex project, such as image classification or a recommendation system. Consider using healthcare datasets if available.

### **Semester 3: Specialized Learning in Medical AI**
- **Coursework:**
  - Choose electives that relate to healthcare technologies or bioinformatics.
  - Consider a course in **ethics** and **data privacy** to understand the implications of AI in healthcare.

- **Skills Development:**
  - Start learning about **medical datasets** (e.g., medical imaging, electronic health records).
  - Gain experience with **cloud platforms** (AWS, Google Cloud) as they are often used in AI projects.

- **Project:**
  - Begin a project focused on medical AI, such as analyzing medical images or creating a predictive model for patient outcomes.

### **Semester 4: Internship and Real-World Experience**
- **Internship:**
  - Apply for internships in companies or research institutions working on medical AI. This will give you practical experience and industry exposure.

- **Skills Development:**
  - Learn about **natural language processing (NLP)** in healthcare, such as processing medical notes or patient reviews.

- **Project:**
  - If you’re not interning, contribute to an open-source medical AI project or collaborate with a professor on research.

### **Semester 5: Portfolio and Freelancing Prep**
- **Coursework:**
  - Take any remaining advanced electives in AI/ML or healthcare-related topics.
  - Consider a course on **project management** or **entrepreneurship** if available.

- **Skills Development:**
  - Focus on mastering **deployment** skills (e.g., Docker, Flask, or FastAPI) to deploy your models in real-world applications.

- **Project:**
  - Work on a capstone project that integrates all your learnings. Aim for a project that is complex, impactful, and relevant to medical AI.

- **Freelancing Prep:**
  - Begin building your online portfolio (GitHub, personal website).
  - Start applying for freelancing opportunities, especially those related to medical AI.

### **Semester 6: Transition to Freelancing and MBBS Prep**
- **Final Preparation:**

- **Freelancing:**
  - By this time, aim to have steady freelancing work or project-based jobs in medical AI. This will help you sustain yourself during your MBBS.

- **Networking:**
  - Attend conferences, seminars, or online webinars related to medical AI. Build a network of professionals in the field.

- **Portfolio Finalization:**
  - Ensure your portfolio is polished, showcasing all your projects and skills. Consider writing blog posts or case studies to highlight your expertise.

### **Additional Tips:**
- **Continuous Learning:** Keep up with the latest trends in AI/ML and medical technology through online courses, tutorials, and research papers.

- **Well-being:** Ensure you have time for self-care and social activities to avoid burnout.
